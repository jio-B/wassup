# Label Encoding
- 문자형 변수를 숫자형 변수로 인코딩

```python
import warnings
warnings.filterwarnings('ignore')
```

### 실습 데이터
- 자동차 연비(mpg) 데이터
    - mpg : miles per gallon
    - 1 mile : 1.6 Km
    - 1 gallon : 3.78 Liter

```python
import seaborn as sns

DF = sns.load_dataset('mpg')
```

```python
DF.info()
```

- 문자형 데이터 : 'origin'

```python
DF.head()
```

```python
type(DF.origin[0])
```

- 명목형 : 이름확인 및 빈도분석

```python
DF.origin.value_counts()
```

- 'origin' Data

```python
X = DF[['origin']]
```

```python
X[111:115]
```

## With LabelEncoder
- 정수(Integer) 인코딩
- (정수로 만들어야 계산이 가능하니까)

```python
from sklearn.preprocessing import LabelEncoder

encoder1 = LabelEncoder()
LE = encoder1.fit_transform(X)
```

- 정수 인코딩 결과

```python
LE[111:115]
```

## With OneHotEncoder
- 원-핫(One-Hot) 인코딩
- 데이터의 양이 비교적 적을 때
- 변수를 동등하게 구분하고 싶을 때(정수라면 영향을 미칠 것 같아서)

```python
from sklearn.preprocessing import OneHotEncoder 

encoder2 = OneHotEncoder()
OHE = encoder2.fit_transform(X)
```

```python
OHE.toarray()[111:115]
```

- Array 변환 필요

```python
print(OHE[111:115])
```

```python
OHE.toarray()[111:115] # 이렇게 봐야 OHE의 값을 알아보기 쉬워진다
```

## Scaling
- 범위(Scale)가 다른 변수의 범위(Scale)를 비슷하게 맞추기 위한 목적
- (정규화 & 표준화와 동일한 절차)

```python
import warnings
warnings.filterwarnings('ignore')
```

- 실습 데이터
    - 1) seaborn 'mpg' Data Set

```python
import seaborn as sns

DF = sns.load_dataset('mpg')
```

- 자동차 연비(mpg) 데이터
    - mpg : miles per gallon
    - 1 mile : 1.6 Km
    - 1 gallon : 3.78 Liter

```python
DF.info()
```

- X, y Data

```python
X = DF[['weight']]
y = DF['mpg']
```

## Without Scaling
- X값 확인

```python
X[:5]
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (9, 6))
sns.regplot(x = X, y = y)
plt.xlabel('weight_Without_Scaling')
plt.show()
```

## With Normalization
- sklearn Package

```python
from sklearn.preprocessing import MinMaxScaler

scaler1 = MinMaxScaler()
X_Norm = scaler1.fit_transform(X)
```

- 정규화된 X값 확인

```python
X_Norm[:5]  # 0~1 사이로 변형
```

- 정규화된 X값 모델생성

```python
fig = plt.figure(figsize = (9, 6))
sns.regplot(x = X_Norm, y = y)
plt.xlabel('weight_With_Normalization')
plt.show()

# 0 ~ 1사이의 값으로 바꿔도(정규화해도) 전체 분포에 영향이 없음
```

## With Standardization
- sklearn Package

```python
from sklearn.preprocessing import StandardScaler

scaler2 = StandardScaler()
X_Stan = scaler2.fit_transform(X)
```

- 표준화된 X값 확인

```python
X_Stan[:5]
```

- 표준화된 X값 모델생성

```python
fig = plt.figure(figsize = (9, 6))
sns.regplot(x = X_Stan, y = y)
plt.xlabel('weigh_With_Standardization')
plt.show()
```
