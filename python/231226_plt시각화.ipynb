## Axes 객체와 subplot 명령
- 하나의 윈도우(Figure)안에 여러개의 플롯을 배열 형태로 봐야하는 경우도 있다.
- Figure 안에 있는 각각의 플롯은 Axes 라고 불리는 객체에 속한다. 
- Axes 객체에 대한 자세한 설명은 다음 웹사이트를 참조한다.
- http://matplotlib.org/api/axes_api.html#matplotlib.axes.Axes

- tight_layout 명령을 실행하면 플롯간의 간격을 자동으로 맞춰준다.

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)   # 여기까지 데이터 생성

ax1 = plt.subplot(2, 1, 1)  # 2행, 1열, 순서(첫번째)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')  # 여기까지 plot 1개 생성

ax2 = plt.subplot(2, 1, 2) # 2행, 1열, 순서(두번째)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped') # 여기까지 또 다른 plot 1개 생성

plt.tight_layout()
plt.show()
```

- 만약 2x2 형태의 네 개의 플롯이라면 다음과 같이 그린다.
- 이 때 subplot 의 인수는 (2,2,1)를 줄여서 221 라는 하나의 숫자로 표시할 수도 있다.
- Axes의 위치는 위에서 부터 아래로, 왼쪽에서 오른쪽으로 카운트한다.

```python
np.random.seed(0)

plt.subplot(221)  # 2행 2열의 첫번째 그림  
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
```

## Axis 객체와 축
- 하나의 Axes 객체는 두 개 이상의 Axis 객체를 가진다.- 
Axis 객체는 플롯의 가로축이나 세로축을 나타내는 객체이다. 보다 자세한 내용은 다음 웹사이트를 참조한다.- 

https://matplotlib.org/api/axis_api.html

### (엄청 중요함) 이중축 twinx
- twinx 명령은 x 축을 공유하는 새로운 Axes 객체를 만든다.

```python
fig, axes1 = plt.subplots()

axes1.set_title("2개의 y축 한 figure에서 사용하기")
axes1.set_xlabel("공유되는 x축")

axes1.plot([10, 5, 2, 9, 7], 'r-') # 기준 주차트
axes1.set_ylabel("y0")

axes2 = axes1.twinx()  # twinx : 레이블, 플롯 셋팅 완료 후 복제할거야 
# 빨간색 선(=기준 주차트)만 복제함

axes2.plot([100, 200, 220, 180, 120], 'g:') # 보조차트
axes2.set_ylabel("y1")

plt.show()
```
