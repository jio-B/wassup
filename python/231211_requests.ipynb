```python
# (전처리) 마지막 더보기 링크 제외 (2)
# 인덱스로 끝에 '더보기'를 잘라버릴까? 

srclist = soup.select('.news_area a')[:-1]
srclist
```

## 상대 URL을 절대 URL로 변환하기

```python
href = sample['href']
```

```python
# //가 중복되는 문제처럼 생각보다 합치는게 번거로울 때가 있음
# '~ finance.naver.com//news ~'
PATH + href
```

```python
# 이럴 땐 urllib 활용 (추천)
from urllib.parse import urljoin
```

```python
# //가 중복되는 문제 없음

urljoin(PATH, href)
```

- 여기까지 '주요뉴스' 6개 리스트 중 1개를 가져오는 절차를 거침.
- 아직 5개의 리스트를 더 가져와야함(절차 동일)
- 이럴 때 반복문을 쓰는 것!!

```python
# 제목과 url 데이터 목록 만들기
news_title= []
news_url = []

for n in srclist: 
    # 태그 객체의 메서드(.txt)를 쓰려면 태그 객체별로 빼줘야함
    title = n.text  # n 요소에서 텍스트만 뽑아줘
    url = urljoin(PATH, n['href'])
    print(title, url) # 제대로 뽑았는지 중간체크
    news_title.append(title) # 리스트에 담아줘1
    news_url.append(url) # 리스트에 담아줘2   
```

```python
news_title  # 타이틀만 가져와줘
```

```python
news_url # 링크주소만 가져와줘
```

# 파라미터 처리하기
- 레스트(REST)란? get과 post를 구분하기(아래 링크 참고)
- https://resttesttest.com//

```python
import requests
```

```python
param = {'name':'Amy', 'age':20, 'address':'Seould'}
```

```python
# 요청 정보(파라미터)가 인자에 포함되어 있다.
resp1 = requests.get('http://httpbin.org/get', params=param) 
# get 방식 먼저 보자
# get 방식은 파라미터 쓸 수 있다

print(resp1)
print(resp1.text)
```

```python
# 요청 정보(파라미터)가 인자에 없고, form에 숨겨져있다.(예시라서 보이는 것)
# post 방식 보자
# post는 파라미터를 숨겨줘야 하므로 'data=param' 으로 변경
# form에 숨겨져 있음

resp2 = requests.post('http://httpbin.org/post', data=param)
print(resp2.text)
```
