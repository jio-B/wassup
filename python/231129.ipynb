## for와 range를 이용한 구구단
- https://wikidocs.net/22

```python
# 이중 for문
for i in range(2, 10):  # 1번 for문
    for j in range(1, 10):  # 2번 for문
        print(i*j, end=" ")
    print()    
```

```python
# x 뽑기
for i in range(2, 10):
    print(i)
```

```python
# x 뽑기 + y뽑기 추가
for i in range(2, 10):
    print('x', i)   # print를 수시로 넣어서(=logging) 프로그램 진행상황을 체크하자
    for j in range(2, 10):
        print('y', j)
        print('-'*50)
```

```python
# x 뽑기 + y뽑기 추가
for i in range(2, 10):
    print(f'==={i}단 시작===')
    print('x', i)   # print를 수시로 넣어서(=logging) 프로그램 진행상황을 체크하자
    for j in range(1, 10):
        print('y', j)
        print(f'{i} * {j} = {i*j}')
        print('-'*50)
```

### 리스트 컴프리헨션 

```python
# 샘플코드

a = [1, 2, 3, 4]
result = []

for num in a:  # in 뒤에는 반복가능한(iterable) 객체가 들어간다
    result.append(num*3)   # 모든 객체에 3씩 곱하기 : 1*3, 2*3, 3*3, 4*3

print(result)
```

```python
# 앞에 거 간단하게 작성하기 (리스트 컴프리헨션 적용하기)
# 리스트 안에서 for문까지 해결 
a = [1, 2, 3, 4]

result = [num * 3 for num in a]
print(result)
```

```python
# 짝수만 담기 : if문 활용
# [표현식 for 항목 in 반복_가능_객체 if 조건문]

a = [1, 2, 3, 4]
result = [num * 3 for num in a if num % 2 ==0]
print(result)
```

```python
# 중첩 리스트 컴프리헨션
result = [x*y for x in range(2,10)
         for y in range(1,10)]
print(result)
```

```python
# 리스트 컴프리헨션 퍼포먼스 비교
import time

start = time.time()
numbers = [i for i in range(1000000)]
end = time.time()
print('List Comprehension Time:', end-start)
```

```python
start = time.time()
numbers = []
for i in range(1000000):
    numbers.append(i)
end = time.time()
print('For Loop Time:', end-start)
```
