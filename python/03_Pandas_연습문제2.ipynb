
````python
df_covid['연번'] = df_covid['연번'].astype(str)
df_covid = df_covid.drop(["연번"], axis=1)
````

# Q1-5
'확진일' 컬럼값을 아래와 같은 형태로 변경하여라
<br>
'1.24.' => '2020-1-24'

````python
from datetime import datetime

df_covid['확진일'] = pd.to_datetime('2020-' + df_covid['확진일'], errors='coerce')
df_covid['확진일'] = df_covid['확진일'].dt.strftime('%Y-%m-%d')
````

````python
df_covid.head()
````
